```python
# create empty column
df['BMI_category'] = ""
```

```python
col1= 'Gender'
col2= 'Age on testday (t0)'
col3= 'BMI'
col4= 'BMI_category'
# BMI syntax for the Boys
def myfunc(col1, col2, col3, col4):
   if col1 ==0 and col2 > 2.00 and col2 <= 4.99 and col3 > 0 and col3 <= 17.54: col4=1
   elif col1==0 and col2 > 2.00 and col2 <= 4.99 and col3 > 17.55 and col3 <= 19.28: col4=2
   elif col1==0 and col2 > 2.00 and col2 <= 4.99 and col3 > 19.29 and col3 <= 35.00: col4=3
   elif col1==0 and col2 > 5.00 and col2 <= 5.99 and col3 > 0 and col3 <= 17.41: col4=1
   elif col1==0 and col2 > 5.00 and col2 <= 5.99 and col3 > 17.42 and col3 <= 19.29: col4=2
   elif col1==0 and col2 > 5.00 and col2 <= 5.99 and col3 > 19.30 and col3 <= 35.00: col4=3 
   elif col1==0 and col2 > 6.00 and col2 <= 6.99 and col3 > 0 and col3 <= 17.54: col4=1
   elif col1==0 and col2 > 6.00 and col2 <= 6.99 and col3 > 17.55 and col3 <= 19.77: col4=2
   elif col1==0 and col2 > 6.00 and col2 <= 6.99 and col3 > 19.78 and col3 <= 35.00: col4=3 
   elif col1==0 and col2 > 7.00 and col2 <= 7.99 and col3 > 0 and col3 <= 17.91: col4=1
   elif col1==0 and col2 > 7.00 and col2 <= 7.99 and col3 > 17.92 and col3 <= 20.62: col4=2 
   elif col1==0 and col2 > 7.00 and col2 <= 7.99 and col3 > 20.63 and col3 <= 35.00: col4=3       
   else: col4=col4
   return col4

df['BMI_category'] = df.apply(lambda x: myfunc(x[col1], x[col2], x[col3], x[col4]), axis=1)
```
